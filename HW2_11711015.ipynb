```python
import urllib.request as req
import json
import pandas as pd
from sklearn import *
```

```python
url = 'https://api.inews.qq.com/newsqa/v1/query/inner/publish/modules/list?modules=chinaDayList,chinaDayAddList,nowConfirmStatis,provinceCompare'
request = req.Request(url)
request.add_header("user-agent","Mozilla/5.0")
response = req.urlopen(request)
json_ret = json.loads(response.read())
data = json_ret['data']
chinaDayAddList = data['chinaDayAddList']
chinaDayList = data['chinaDayList']
```

```python
df1 = pd.DataFrame(chinaDayAddList)
df2 = pd.DataFrame(chinaDayList)
df1 = df1[df1['date'] > '03.14']
df1 = df1[df1['date'] < '04.14']
df2 = df2[df2['date'] > '03.14']
df2 = df2[df2['date'] < '04.14']
df2
```

```python
acc_df = df2[['local_acc_confirm', 'nowConfirm', 'heal', 'dead', 'date']]
add_df = df1[['localConfirmadd', 'suspect', 'date']]
```

```python
acc_df.columns = ['acc_confirm', 'confirm', 'acc_heal', 'acc_dead', 'date']
add_df.columns = ['confirm_add', 'suspect_add', 'date']
df = pd.merge(acc_df, add_df, 'left')
df
```

```python
df['suspect_add'].fillna(0)
df['confirm_add'].fillna(df['confirm_add'].mean())
df
```

```python
method = {'confirm_add' : df['confirm_add'].mean(), 'suspect_add' : 0}
df  = pd.DataFrame.fillna(df, method)
df.to_dict()
```

```python
train_df = df.iloc[0 : 24]
train_df = train_df.copy()
train_df['date'] = [i for i in range(1, 25)]
test_df = df.iloc[24:]
test_df = test_df.copy()
test_df['date'] = [i for i in range(25, 31)]
test_df
```

```python
X_train = train_df['date'].values.reshape(-1,1)
Y_train = train_df['acc_confirm'].values.reshape(-1,1)
X_test = test_df['date'].values.reshape(-1,1)
Y_test = test_df['acc_confirm'].values.reshape(-1,1)
model = linear_model.LinearRegression()
model.fit(X_train, Y_train)
print("intercept", model.intercept_)
print("coef", model.coef_)
Y_prid = model.predict(X_test)
print('rmse', metrics.mean_squared_error(Y_test, Y_prid))
final_pred = model.predict([[31]])
print("final predict", final_pred)
```
